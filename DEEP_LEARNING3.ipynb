```python
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))
def coding_two_layered_single_neurons(x1,w1,w2,b1,b2):
    haidden_layer2_input=x1*w1+b1
    result=sigmoid(haidden_layer2_input)
    exact_result=result*w2+b2
    rr=sigmoid(exact_result)
    return rr
```

```python
x1=0.7
w1=0.2
w2=0.66
b1=0.3
b2=0.7
coding_two_layered_single_neurons(x1,w1,w2,b1,b2)
```
