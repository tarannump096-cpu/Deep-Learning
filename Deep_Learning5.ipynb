```python
import sympy as sp
```

```python
yi=sp.symbols("yi")
w=sp.symbols("w")
x=sp.symbols("x")
b=sp.symbols("b")
```

```python
f=(yi-(w*x+b))**2
```

```python
sp.diff(f,w)
```
