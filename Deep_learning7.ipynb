```python
def SGD_optimizers(x,w,b,lr,la,yi):
    for i in range(100):
        y_pred=x*w+b
        loss=(yi-y_pred)**2

        grad1=(-2*x)*(yi-(w*x+b))
        grad2=2*la*w
        w=w-(lr)*(grad1+grad2)
    return w
```

```python
x=2
w=0.3
b=1
lr=0.00001
la=1
yi=10
SGD_optimizers(x,w,b,lr,la,yi)
```

```python
import numpy as np
def weight_updation_multiple_inputs(x,w,b,lr,la,yi):
    x=0
    for i in range(10000):
        yi_hat=np.dot(x,w)+b
        loss=(yi-yi_hat)**2
        grad1=(-2*x)*(yi-(w*x+b))
        grad2=2*la*w
        w=w-(lr)*(grad1+grad2)
        lr=lr/(lr+x)
        x+1
    return w
```

```python
x=np.array([2,3,4])
w=np.array([0.3,0.2,0.4])
b=1
lr=0.00000001
la=1
yi=10
weight_updation_multiple_inputs(x,w,b,lr,la,yi)
```

```python
def two_weigts_two_layers(x,w1,w2,b1,b2,lr,la,yi):
    x=0
    for i in range(1000):
        o1=(x*w1)+b1
        y_pred=(o1*w2)+b2
        loss=(yi-y_pred)**2
        grad_w1=(-2*x)*(yi-(w1*x+b1))
        grad_w2=(-2*x)*(yi-(w2*x+b2))
        gradR_w1=2*la*w1
        gradR_w2=2*la*w2
        w1=w1-(lr)*(grad_w1+gradR_w1)
        w2=w2-(lr)*(grad_w2+gradR_w2)
        lr=lr/(lr+x)
        x+=1
    return w1,w2
```

```python
x=2
w1=0.3
w2=0.5
b1=b2=1
lr=0.00001
la=1
yi=10
two_weigts_two_layers(x,w1,w2,b1,b2,lr,la,yi)
```
