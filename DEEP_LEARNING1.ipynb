```python
def coding_single_neuron(inputs,weights,bias):
    output=0
    for i in range(len(inputs)):
        res=inputs[i]*weights[i]
        output+=res
    output+=bias
    return output
```

```python
inputs=[2.1,3.4,2.0,1.0,3.3]
weights=[2,3,3.2,4,5]
bias=0.4
output=coding_single_neuron(inputs,weights,bias)
```

```python
import numpy as np
def coding_single_neuron_numpy(inputs,weights,bias):
    input_array=np.array(inputs)
    weights_array=np.array(weights)
    output=input_array*weights_array+bias
    return np.sum(output)
```
