```python
import numpy as np
def sigmoid(x):
    return 1/(1+np.exp(-x))
def train_single_hidden_layer_NN(inputs,weights,bias1,bias2):
    o1=sigmoid(inputs[0]*weights[0])+bias1
    o2=sigmoid(inputs[1]*weights[1])+bias2
    o=sigmoid(o1*weights[2]+o2*weights[3])
    return o
```

```python
inputs=[0.8,0.9]
weights=[0.65,0.78,0.66,0.45]
bias1=0.88
bias2=0.87
yi_hat=train_single_hidden_layer_NN(inputs,weights,bias1,bias2)
```
