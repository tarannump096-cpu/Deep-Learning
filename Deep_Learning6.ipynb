```python
def sigmoid(x):
    return 1/1+np.exp(-x)
```

```python
import numpy as np
def contructing_layer(inputs,weights,bias):
    output=np.dot(weights,inputs)
    output=output+bias
    ans=sigmoid(output)

    return ans
```

```python
inputs=np.array([3,4,5,6])
weights=np.array([[1,2,3,4],
                 [4,5,6,7],
                 [6,7,8,9]])
bias=np.array([1,1,2])
contructing_layer(inputs,weights,bias)
```

```python
def coding_layers_batch(inputs,weights,bias):
    output=np.dot(inputs,weights.T)
    output=output+bias
    return output
```

```python
inputs=np.array([[1,2,3,4],
                  [4,5,6,7],
				  [2,3,6,7]])
weights=np.array([[1,2,3,4],
			     [4,5,6,7],
                 [6,7,8,9]])
bias=np.ones((3,3))
coding_layers_batch(inputs,weights,bias)
```

```python
def coding_two_layers(inputs,weights_layer1,weights_layer2,bias_layer1,bias_layer2):
    o1=np.dot(weights_layer1,inputs)
    o1=o1+bias_layer1
    o2=np.dot(weights_layer2.T,o1)
    o2=o2+bias_layer2
    return o2
```

```python
inputs=np.array([3,4,5,6])
weights_layer1=np.array([[1,2,3,4],
                         [4,5,6,7],
                         [6,7,8,9]])
weights_layer2=np.array([[1,2],
                        [1,3],
                        [1,1]])

bias_layer1=np.array([1,1,2])
bias_layer2=np.array([2,3])
coding_two_layers(inputs,weights_layer1,weights_layer2,bias_layer1,bias_layer2)
```
